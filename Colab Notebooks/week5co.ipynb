```python
import sklearn
import pandas as pd
from sklearn.datasets import load_iris
iris=load_iris()
print(iris.keys())
```

```python
df=pd.DataFrame(iris['data'])
print(df)
print(iris['target_names'])
print(iris['feature_names'])
print(iris['target'])
```

```python
x=df
y=iris['target']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
y_pred=knn.predict(x_test)
cm=confusion_matrix(y_test,y_pred)
print(y_test)
print(y_pred)
print(cm)
```

```python
print("Correct prediction",accuracy_score(y_test,y_pred))
print("Wrong prediction",(1-accuracy_score(y_test,y_pred)))
y_testtrain=knn.predict(x_train)
cm1=confusion_matrix(y_train,y_testtrain)
print(y_train)
print(y_testtrain)
print(cm1)
```

```python
cm2=pd.DataFrame(data=cm,index=['setosa','versicolor','virginica'],columns=['setosa','versicolor','virginica'])
print(cm2)
from sklearn.tree import DecisionTreeClassifier
svc=DecisionTreeClassifier()
svc.fit(x_train,y_train)
y_p=svc.predict(x_test)
print(accuracy_score(y_p,y_test))
```
